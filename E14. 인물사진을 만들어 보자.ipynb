```python
class DeepLabModel(object):
    INPUT_TENSOR_NAME = 'ImageTensor:0'
    OUTPUT_TENSOR_NAME = 'SemanticPredictions:0'
    INPUT_SIZE = 513
    FROZEN_GRAPH_NAME = 'frozen_inference_graph'

    def __init__(self, tarball_path):
        self.graph = tf.Graph()
        graph_def = None
        tar_file = tarfile.open(tarball_path)
        for tar_info in tar_file.getmembers():
            if self.FROZEN_GRAPH_NAME in os.path.basename(tar_info.name):
                file_handle = tar_file.extractfile(tar_info)
                graph_def = tf.compat.v1.GraphDef.FromString(file_handle.read())
                break
        tar_file.close()

        with self.graph.as_default():
            tf.compat.v1.import_graph_def(graph_def, name='')

        self.sess = tf.compat.v1.Session(graph=self.graph)

    def preprocess(self, img_orig):
        height, width = img_orig.shape[:2]
        resize_ratio = 1.0 * self.INPUT_SIZE / max(width, height)
        target_size = (int(resize_ratio * width), int(resize_ratio * height))
        resized_image = cv2.resize(img_orig, target_size)
        resized_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
        img_input = resized_rgb
        return img_input
        
    def run(self, image):
        img_input = self.preprocess(image)

        batch_seg_map = self.sess.run(
            self.OUTPUT_TENSOR_NAME,
            feed_dict={self.INPUT_TENSOR_NAME: [img_input]})

        seg_map = batch_seg_map[0]
        return cv2.cvtColor(img_input, cv2.COLOR_RGB2BGR), seg_map
```

##### 1. 인물검출 시  일부 이미지의 소실과 관련하여 엣지검출을 통한 보정방안

step2의 2, 3에 상술한 바와 같이 가장자리 일부 이미지의 소실과 관련하여 엣지검출을 통하여 보정을 할 수 있을 것이다. 이때 사용되는 필터로는 LoG필터를 이용한다. 이 과정은 크게 3단계로 나누어 진다.

1.1 원본영상에 LoG필터를 이용하여 사진의 엣지를 검출한다.  

1.2 세그멘테이션 결과값과 1.1 결과와 and연산을 통해 1차적으로 자를 부분을 확정한다.  
* 만일 1.2의 결과만 놓고 본다면, 해당 결과 값은 세그멘테이션 결과와 같을 것이다.

1.3 세그멘테이션 결과값과 1.1 결과와 or연산을 통하여 보정할 부분의 구획을 나누어 준다.
* 이때 엣지로 구획을 나누어 준 이유는 엣지 내부의 값은 유사할 것이라 추정
* 예를 들어 step2 3번 사진 중 어깨부분의 blur처리 된 것과 관련 원본 영상의 경우 어깨부분의 엣지를 추출 할 수 있고 엣지 외부를 검은색 의자라 가정한다면, 마이와 의자의 경계선의 픽셀의 값이 급격하게 차이가 날 것이다.
* 따라서 시작값과 상하좌우를 비교하여 값이 유사하다면 엣지까지 같은 값으로 채움으로 연산량을 줄인다.
